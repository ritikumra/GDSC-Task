## **Load Data and Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df= pd.read_csv('https://raw.githubusercontent.com/ritikumra/GDSC-Task/main/laptopData.csv')
  
  
```

```python
df=df.dropna()
```

```python
y = df[ 'Amount']
y
```

```python
x = df.drop(['S_No','Amount'],axis=1)
x
```

```python
x
```

### Removing Units and changing datatypes


```python
x["RAM"] = x["RAM"].str.replace("GB", "")
x["RAM"] = x["RAM"].astype(int)
```

```python
x["Mass(in kg)"] = x["Mass(in kg)"].str.replace("kg", "")
x["Mass(in kg)"] = x["Mass(in kg)"].str.replace("?", "0")
x["Mass(in kg)"] = x["Mass(in kg)"].astype(float)

```

```python
x
```

###Splitting Screen Resolution into X and Y and then multiplying

```python
new = x["Screen Resolution"].str.split("x", n = 1, expand = True) 
x["X_res"]= new[0]
x["Y_res"]= new[1]
x["Y_res"]= pd.to_numeric(x["Y_res"])
x["Y_res"]= x["Y_res"].astype(float)
x["X_res"]=(x['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x: pd.Series(x).astype(int)).mean(1))
x["X_res"]=pd.to_numeric(x["X_res"])
x["Screen Resolution"]=(x["X_res"]*x["Y_res"]).astype(float)
```

```python
x.drop(['X_res','Y_res'],axis =1 , inplace = True)
```

```python
x
```

###Splitting Memory column into HDD nd SSD

```python
x['Memory'] = x['Memory'].astype(str).replace('\.0', '', regex=True)
x["Memory"] = x["Memory"].str.replace('GB', '')
x["Memory"] = x["Memory"].str.replace('TB', '000')
new2 = x["Memory"].str.split("+", n = 1, expand = True)
x["first"]= new2[0]
x["first"]=x["first"].str.strip()
x["second"]= new2[1]
x["Layer1HDD"] = x["first"].apply(lambda x: 1 if "HDD" in x else 0)
x["Layer1SSD"] = x["first"].apply(lambda x: 1 if "SSD" in x else 0)
x["Layer1Hybrid"] = x["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
x["Layer1Flash_Storage"] = x["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)
x['first'] = x['first'].str.replace(r'\D', '')
x["second"].fillna("0", inplace = True)
x["Layer2HDD"] = x["second"].apply(lambda x: 1 if "HDD" in x else 0)
x["Layer2SSD"] = x["second"].apply(lambda x: 1 if "SSD" in x else 0)
x["Layer2Hybrid"] = x["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
x["Layer2Flash_Storage"] = x["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)
x['second'] = x['second'].str.replace(r'\D', '')
```

```python
x['first'] = pd.to_numeric(x['first'])
x["second"] = x["second"].astype(int)
x["Total_Memory"]=(x["first"]*(x["Layer1HDD"]+x["Layer1SSD"]+x["Layer1Hybrid"]+x["Layer1Flash_Storage"])+x["second"]*(x["Layer2HDD"]+x["Layer2SSD"]+x["Layer2Hybrid"]+x["Layer2Flash_Storage"]))
x["Memory"]=x["Total_Memory"]
x["HDD"]=(x["first"]*x["Layer1HDD"]+x["second"]*x["Layer2HDD"])
x["SSD"]=(x["first"]*x["Layer1SSD"]+x["second"]*x["Layer2SSD"])
x["Hybrid"]=(x["first"]*x["Layer1Hybrid"]+x["second"]*x["Layer2Hybrid"])
x["Flash_Storage"]=(x["first"]*x["Layer1Flash_Storage"]+x["second"]*x["Layer2Flash_Storage"])
```

```python
x.drop(['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid','Layer1Flash_Storage','Layer2HDD','Layer2SSD','Layer2Hybrid','Layer2Flash_Storage','Total_Memory'],axis=1,inplace = True)
```
